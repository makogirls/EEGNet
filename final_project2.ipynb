```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=corr,annot=True,
fmt='.2f',linewidths=.5,cmap='Blues')
#상관계수 시각화
```
